```python
td_raw = str(td_raw)
print(td_raw)
```

```python
data_index = td_raw.find("data")
```

```python
td_raw = td_raw[data_index + 8: -23]
```

```python
td_raw = td_raw.split('],[')
print(td_raw[0])
print(len(td_raw)) 
temp = '[' + td_raw[0] + ']'
temp_list = json.loads(temp)
print(temp_list)
print(type(temp_list))
```

```python
res = [];
for i in range(len(td_raw)):
    temp = '[' + td_raw[i] + ']';
    temp_list = json.loads(temp);
    res.append(temp_list);
```

```python
tdraw_df = pd.DataFrame(res) 
tdraw_df.to_excel("td_raw.xlsx")
```

```python
td_clean = tdraw_df.rename(columns={0: 'UTC', 1:'TotalDemand'});
split_data = td_clean['UTC'].str.split("T", n = 1, expand = True)
td_clean["Date"] = split_data[0];
td_clean["Time"] = split_data[1].str.split("Z", n = 1, expand = True)[0]
td_clean.drop(columns =["UTC"], inplace = True)
td_clean.to_excel("td_clean.xlsx")
```
