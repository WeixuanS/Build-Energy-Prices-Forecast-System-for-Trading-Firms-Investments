```
html = urlopen('http://api.eia.gov/series/?api_key=bdc665fdd718f4f2624518fada5e39f8&series_id=EBA.US48-ALL.DF.H')
```

```
fd_raw = BeautifulSoup(html.read(), "lxml")
```

```
fout = open('fdraw_temp.txt', 'wt', encoding='utf-8')
fout.write(str(fd_raw))
fout.close()
```

```
fd_raw = str(fd_raw)
print(fd_raw)
```

```
data_index = fd_raw.find("data")
```

```
fd_raw = fd_raw[data_index + 8: -23]
```

```
fd_raw = fd_raw.split('],[')
print(fd_raw[0])
print(len(fd_raw)) 
temp = '[' + fd_raw[0] + ']'
temp_list = json.loads(temp)
print(temp_list)
print(type(temp_list))
```

```
res = [];
for i in range(len(fd_raw)):
    temp = '[' + fd_raw[i] + ']';
    temp_list = json.loads(temp);
    res.append(temp_list);
```

```
fdraw_df = pd.DataFrame(res) 
fdraw_df.to_excel("fd_raw.xlsx")
```

```
fd_clean = fdraw_df.rename(columns={0: 'UTC', 1:'forecast'});
split_data = fd_clean['UTC'].str.split("T", n = 1, expand = True)
fd_clean["Date"] = split_data[0];
fd_clean["Time"] = split_data[1].str.split("Z", n = 1, expand = True)[0]
fd_clean.drop(columns =["UTC"], inplace = True)
fd_clean.to_excel("fd_clean.xlsx")
```
